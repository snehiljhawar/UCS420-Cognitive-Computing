Q1: Implement logistic regression on the Iris dataset.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

```python
df=datasets.load_iris()
df
```

```python
X=df['data']
X
```

```python
X=pd.DataFrame(X,columns=['sepal_length','sepal_width','petal_length','petal_width'])
```

```python
X.head()
```

```python
Y=df['target']
Y
```

```python
X.shape,Y.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
model=LogisticRegression()
model.fit(X_train, Y_train)
```

```python
train_pred=model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
train_pred
```

```python
test_pred
```

```python
train_Score=accuracy_score(train_pred,Y_train)
print("Training Accuracy:",train_Score)

test_Score=accuracy_score(test_pred,Y_test)
print("Testing Accuracy:",test_Score)
```

```python
train_matrix=confusion_matrix(train_pred,Y_train)
print("Training matrix:\n",train_matrix)

test_matrix=confusion_matrix(test_pred,Y_test)
print("Testing matrix:\n",test_matrix)
```

```python
train_report=classification_report(train_pred,Y_train)
print("Training report:",train_report)

test_report=classification_report(test_pred,Y_test)
print("Testing report:",test_report)
```

```python
import warnings
warnings.filterwarnings("ignore")

input_data=(['4.9', '3' , '1.4',' 0.2'])

scaled_data=scaler.transform([input_data])

prediction=model.predict(scaled_data)[0]
print("Predicted value is : ",prediction)
```
