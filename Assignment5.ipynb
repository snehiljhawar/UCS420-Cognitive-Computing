Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find<br>
i. Sum of all elements<br>
ii. Sum of all elements row-wise<br>
iii. Sum of all elements column-wise

```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

total_sum = np.sum(gfg)
print("Sum of all elements:", total_sum)

row_wise_sum = np.sum(gfg, axis=1)
print("Row-wise sum:\n", row_wise_sum)

column_wise_sum = np.sum(gfg, axis=0)
print("Column-wise sum:", column_wise_sum)
```

Q.2 (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find<br>
i. Sorted array<br>
ii. Indices of sorted array<br>
iii. 4 smallest elements<br>
iv. 5 largest elements

```python
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array= np.sort(array)
print("Sorted array: ", sorted_array)

sorted_indices = np.argsort(array)
print("Indices of sorted array:", sorted_indices)

smallest_4 = np.sort(array)[:4]
print("4 smallest elements:", smallest_4)

largest_5 = np.sort(array)[-5:]
print("5 largest elements:", largest_5)

```

(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find<br>
i. Integer elements only<br>
ii. Float elements only

```python
import numpy as np
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array[array == array.astype(int)]
print("Integer elements:", integer_elements)

float_elements= array[array!=array.astype(int)]
print("Float elements:", float_elements)

```

Q.3 You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.


a) Generate your unique sales dataset:  Take the sum of the ASCII values of the initials of your first and last name. Call this value X. (If your initials are A B → ASCII sum = 65

66 = 131 → sales = [131, 181, 231, 281, 331].)  Create a NumPy array sales with values [X, X+50, X+100, X+150, X+200].


```python
import numpy as np

first_initial = 'S'
last_initial = 'J'

X = ord(first_initial) + ord(last_initial)
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales array:", sales)
```

b) Compute your personalized tax rate as ((X % 5) + 5) / 100.
 Use broadcasting to apply this tax rate to each sales value.

```python
tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 + tax_rate)
print("Sales after tax:", sales_after_tax)
```

c) Adjust sales based on discount:
 If sales < X+100, apply a 5% discount.
 If sales >= X+100, apply a 10% discount.


```python
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Discounted sales:", discounted_sales)
```

d) Expand sales data for multiple weeks:
 Create a 3×5 matrix representing three weeks of sales by stacking
sales three times using broadcasting.
 Increase sales by 2% per week using element-wise broadcasting.

```python
weekly_sales = np.tile(discounted_sales, (3, 1))
week_factors = np.array([1.00, 1.02, 1.04]).reshape(3, 1)
adjusted_weekly_sales = weekly_sales * week_factors

print("Weekly sales after 2% increase per week:\n", adjusted_weekly_sales)
```

Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use
each function from the list below and compute y values using NumPy:<br>
 Y = x2<br>
 Y = sin(x)<br>
 Y = ex<br>
 Y = log(|x| + 1)<br>
Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 100)


y1 = x ** 2                 
y2 = np.sin(x)               
y3 = np.exp(x)                
y4 = np.log(np.abs(x) + 1)    


plt.figure(figsize=(10, 6))


plt.subplot(2, 2, 1)
plt.plot(x, y1, color='r', label=r'$y = x^2$')
plt.title("Quadratic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()


plt.subplot(2, 2, 2)
plt.plot(x, y2, color='b', label=r'$y = \sin(x)$')
plt.title("Sine Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()


plt.subplot(2, 2, 3)
plt.plot(x, y3, color='g', label=r'$y = e^x$')
plt.title("Exponential Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()


plt.subplot(2, 2, 4)
plt.plot(x, y4, color='purple', label=r'$y = \log(|x| + 1)$')
plt.title("Logarithmic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()


plt.tight_layout()
plt.show()
```
