```python
import numpy as np
import pandas as pd
r_no = 102317171
np.random.seed(r_no)

sales_data = np.random.randint(1000, 5000, size=(12, 4))

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

sales = pd.DataFrame(sales_data, index=months, columns=columns)

print(sales)
```

```python
print(sales.head())
print()
print(sales.describe())

total_sales_per_category = sales.sum()
print()
print(total_sales_per_category)

total_sales_per_month = sales.sum(axis=1)
print()
print(total_sales_per_month)

average_growth = sales.pct_change().mean()
print()
print(average_growth)

sales["Total Sales"] = total_sales_per_month
sales["Growth Rate"] = sales["Total Sales"].pct_change() * 100

if(r_no%2==0):
  sales["Electronics"] *= 0.9
else:
  sales["Clothing"] *= 0.85
print()
print(sales)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=sales[columns], markers=True)
plt.title("Monthly Sales Trends for Each Category")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(columns)
plt.grid(True)
plt.show()
print()
print()
plt.figure(figsize=(10, 6))
sns.boxplot(data=sales[columns])
plt.title("Sales Distribution for Each Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])


abs_array = np.abs(array)

flat_percentiles = np.percentile(array, [25, 50, 75])  
col_percentiles = np.percentile(array, [25, 50, 75], axis=0)  
row_percentiles = np.percentile(array, [25, 50, 75], axis=1)  


flat_mean = np.mean(array)
flat_median = np.median(array)
flat_std = np.std(array)

col_mean = np.mean(array, axis=0)
col_median = np.median(array, axis=0)
col_std = np.std(array, axis=0)

row_mean = np.mean(array, axis=1)
row_median = np.median(array, axis=1)
row_std = np.std(array, axis=1)


print("Element-wise Absolute Value:\n", abs_array, "\n")

print(f"Flattened Array Percentiles (25th, 50th, 75th):\n{flat_percentiles}\n")
print(f"Column-wise Percentiles:\n{col_percentiles}\n")
print(f"Row-wise Percentiles:\n{row_percentiles}\n")

print(f"Flattened Array: Mean={flat_mean}, Median={flat_median}, Std Dev={flat_std}\n")
print(f"Column-wise: Mean={col_mean}, Median={col_median}, Std Dev={col_std}\n")
print(f"Row-wise: Mean={row_mean}, Median={row_median}, Std Dev={row_std}")
```

```python
def swap(lst, index1, index2):
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp
    return lst

my_list = [1, 2, 3, 4, 5]
index1, index2 = 2, 3
print("Original List:", my_list)
print("Swapped List:", swap(my_list, index1, index2))
```
