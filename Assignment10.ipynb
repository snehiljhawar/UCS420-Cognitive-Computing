Q2. Using the same paragraph from Q1:
1. Extract all words with only alphabets using re.findall()
2. Remove stop words using NLTK’s stopword list
3. Perform stemming with PorterStemmer
4. Perform lemmaƟzaƟon with WordNetLemmaƟzer
5. Compare the stemmed and lemmaƟzed outputs and explain when you’d prefer one over
the other. 


```python
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import WordPunctTokenizer
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt

word_tokenizer = WordPunctTokenizer()

paragraph = """My favorite book is a gripping mystery thriller filled with unexpected twists and turns. 
The story revolves around a detective who unravels a complex web of lies in a small town. 
Each chapter ends with a cliffhanger, making it impossible to stop reading. 
The plot is both clever and chilling, keeping readers on edge from start to finish. 
I love how the author blends suspense with deep character development. 
It's the kind of fiction that stays with you long after you turn the final page."""

words_only_alpha = re.findall(r'\b[a-zA-Z]+\b', paragraph)

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_only_alpha if word.lower() not in stop_words]
porter_stemmer = PorterStemmer()
stemmed_words = [porter_stemmer.stem(word) for word in filtered_words]

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\nOriginal Words (Only Alphabets):", words_only_alpha)
print("\nFiltered Words (No Stopwords):", filtered_words)
print("\nStemmed Words (PorterStemmer):", stemmed_words)
print("\nLemmatized Words (WordNetLemmatizer):", lemmatized_words)

plt.figure(figsize=(10, 6))
plt.bar(filtered_words, [len(word) for word in filtered_words], alpha=0.5, label='Filtered Words Length')
plt.bar(stemmed_words, [len(word) for word in stemmed_words], alpha=0.7, label='Stemmed Words Length')
plt.bar(lemmatized_words, [len(word) for word in lemmatized_words], alpha=0.3, label='Lemmatized Words Length')
plt.xticks(rotation=45)
plt.legend()
plt.title("Comparison of Filtered, Stemmed, and Lemmatized Words")
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
texts = [
    "Stock Market Hits New Highs Amid Economic Growth",
    "Tech Industry Faces Challenges with Supply Chain Issues",
    "Government Introduces New Policy for Climate Change Action"
]
count_vectorizer = CountVectorizer()
bow_matrix = count_vectorizer.fit_transform(texts)
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(texts)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

print("Bag of Words Representation (CountVectorizer):")
print(bow_df)

print("\nTF-IDF Representation (TfidfVectorizer):")
print(tfidf_df)

print("\nTop 3 Keywords from Each Document (based on TF-IDF):")
for i, text in enumerate(texts):
    print(f"\nText {i+1}:")
    tfidf_scores = dict(zip(tfidf_vectorizer.get_feature_names_out(), tfidf_matrix[i].toarray()[0]))
    sorted_tfidf = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:3]
    print(sorted_tfidf)


```

```python
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

text1 = """Artificial Intelligence (AI) enables machines to mimic human intelligence.
It powers applications like chatbots, image recognition, and autonomous vehicles.
AI relies on algorithms and massive data for decision-making and learning.
It is transforming industries by automating tasks and predicting outcomes."""

text2 = """Blockchain is a decentralized technology that records data securely.
It underlies cryptocurrencies like Bitcoin and Ethereum.
Blockchain ensures transparency and trust by using cryptographic techniques.
It is revolutionizing finance, supply chains, and digital identity management."""

def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    return tokens

tokens1 = preprocess(text1)
tokens2 = preprocess(text2)

def jaccard_similarity(set1, set2):
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union)

jaccard = jaccard_similarity(set(tokens1), set(tokens2))
print("Jaccard Similarity:", round(jaccard, 4))

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([text1, text2])
cos_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
print("Cosine Similarity:", round(cos_sim, 4))

print("\n Analysis: ")
if jaccard < cos_sim:
    print("Cosine Similarity provides better insight in this case.")
    print("It considers context and word frequency, capturing topic similarity more accurately.")
else:
    print("Jaccard Similarity provides better insight here due to high word overlap.")
```

```python
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt

reviews = [
    "I absolutely love this phone. The battery life is amazing and the camera is outstanding!",
    "The service was okay, not too bad but nothing special either.",
    "Terrible experience! The product broke after two days. Completely disappointed."
]

positive_reviews_text = ""

print(" Review Sentiment Analysis: ")
for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity

    if polarity > 0.1:
        sentiment = "Positive"
        positive_reviews_text += review + " "
    elif polarity < -0.1:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    print(f"\nReview: {review}")
    print(f"Polarity: {polarity:.2f}, Subjectivity: {subjectivity:.2f}")
    print(f"Sentiment: {sentiment}")

if positive_reviews_text.strip():
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive_reviews_text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title("Word Cloud for Positive Reviews")
    plt.show()
else:
    print("\nNo positive reviews to display in the word cloud.")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import random

text = ("Harry Potter is a young wizard who discovers his magical heritage on his eleventh birthday. "
        "He attends Hogwarts School of Witchcraft and Wizardry, where he makes lifelong friends and uncovers "
        "hidden secrets about his past. With courage, loyalty, and the help of his friends, Harry faces dark forces "
        "that threaten the wizarding world. Magic becomes both a gift and a challenge as he learns about his destiny and the power of love.")

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1

input_sequences = []
words = text.lower().split()

for i in range(1, len(words)):
    n_gram_sequence = words[:i+1]
    token_list = tokenizer.texts_to_sequences([" ".join(n_gram_sequence)])[0]
    input_sequences.append(token_list)

max_sequence_len = max([len(seq) for seq in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))

X = input_sequences[:, :-1]
y = tf.keras.utils.to_categorical(input_sequences[:, -1], num_classes=total_words)

model = Sequential()
model.add(Embedding(total_words, 10, input_length=max_sequence_len - 1))  
model.add(LSTM(100))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, verbose=0)  

def generate_text(seed_text, next_words=3):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        
        predicted_probs = model.predict(token_list, verbose=0)
        predicted = np.argmax(predicted_probs, axis=-1)[0]  
        
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                seed_text += " " + word
                break
    return seed_text

seed_word = "magic"
generated = generate_text(seed_word, next_words=3)
print("Generated Text:\n", generated)
```
